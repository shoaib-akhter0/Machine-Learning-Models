```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing Values:\n", df.isnull().sum().sum())
```

```python
# class distribution
class_counts = df[['class_e', 'class_p']].sum()

print(class_counts)

class_counts.plot(kind='bar')
plt.title("Class Distribution (Edible vs Poisonous)")
plt.xticks(rotation=0)
plt.show()
```

```python
corr = df.corr()['class_p'].sort_values(ascending=False)

print("Top 10 positively correlated with Poisonous:\n")
print(corr.head(10))

print("\nTop 10 negatively correlated:\n")
print(corr.tail(10))
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
top_features = corr.index[1:6]  # skip class_p itself

for col in top_features:
    plt.figure()
    sns.boxplot(x=df['class_p'], y=df[col])
    plt.title(f"{col} vs Class")
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop(['class_e', 'class_p'], axis=1)
y = df['class_p']

model = RandomForestClassifier()
model.fit(X, y)

importance = pd.Series(model.feature_importances_, index=X.columns)
importance = importance.sort_values(ascending=False)

print(importance.head(10))
```

```python
importance.head(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Probability of poisonous when feature = 1
prob = df.groupby('class_p').mean().T
print(prob.sort_values(by=1, ascending=False).head(10))
```

```python
X = df.drop("class_e", axis=1)  # depends on encoding
y = df["class_e"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = svm.SVC()
model.fit(X_train, y_train)
```

```python
print("Accuracy:", model.score(X_test, y_test))
```
